# Data Manipulation with Pandas:

## Create a sample dataset using a dictionary

```python
employee_data = {
    "EmployeeID": [101, 102, 103, 104],
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Department": ["HR", "Engineering", "Marketing", "Finance"],
    "Salary": [60000, 80000, 75000, 70000]
}
employee_data
```

## Convert it into a Pandas DataFrame and display it

```python
import pandas as pd
df = pd.DataFrame(employee_data)
df
```

The data above is made up of 4 rows and 4 columns with particulars of the 4 employees captured under the columns **EmployeeID,	Name,	Department, and Salary**

# Data Visualization with Matplotlib/Seaborn

## Create a bar chart or line plot to visualize your data. and labels and titles

```python

import matplotlib.pyplot as plt

df.plot(kind='bar', x='Name', y='Salary')
plt.title('Employee Salaries')
plt.xlabel('Employee')
plt.ylabel('Salary')
plt.show()
```

This bar graph is a visual represantation of the salaries earned by the the 4 employees
