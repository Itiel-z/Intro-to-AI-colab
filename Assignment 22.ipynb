```python
import json


# Write a Python script to read the assignment_data.json file and parse its content.
with open('assignment_data.json') as file:
    data = json.load(file)

# Loop through each user and print their details
for user in data["users"]:
    print(f"User ID: {user['id']}")
    print(f"Name: {user['name']}")
    print(f"Email: {user['email']}")
    print(f"Address: {user['address']['street']}, {user['address']['city']}, {user['address']['state']} {user['address']['zipcode']}")


    print("Orders:")
    for order in user["orders"]:
        print(f"  Order ID: {order['order_id']}, Product: {order['product']}, Amount: ${order['amount']}, Date: {order['date']}")

    print("\n" )
```

**Ensure all extracted email addresses are valid (contain "@" and "."). Convert order amounts to integers (rounded to the nearest whole number).**

```python
# Load the JSON data
with open("assignment_data.json", "r") as file:
    data = json.load(file)

# Loop through the file
for user in data["users"]:
    #name = user["name"]
    #email = user["email"]

    # Validate email
    if "@" in user['email'] and "." in user['email']:
        print(f"Name: {user['name']}, Email: {user['email']}")
    else:
        print(f"Name: {user['name']}, Email: INVALID")

    # City and State
    print(f"City: {user['address']['city']}, State: {user['address']['state']}")

    # Print product names and rounded amounts
    print("Orders:")
    for order in user["orders"]:
        print(f"  Product: {order['product']}, Amount: ${int(round(order['amount']))}")

    print("\n")
```
