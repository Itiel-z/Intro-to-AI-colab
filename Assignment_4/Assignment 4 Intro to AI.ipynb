# Task 3: Handling Null Values

* Create a DataFrame with at least one null value in each column.
* Fill all null values with an appropriate value (e.g., 0 for numeric columns).
* Demonstrate understanding of different methods to handle null values.

```python
# Defining the data
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', '', 'Frank', 'Grace', 'Helen', 'Ian', 'Jack'],
    'Math':    [88, 92, None, 79, 85, None, 90, 76, None, 84],
    'English': [None, 85, 78, 92, None, 80, 87, None, 90, 88],
    'Science': [91, None, 84, None, 89, 76, None, 85, 83, None],
    'History': [87, 90, None, 80, None, None, 88, 92, 75, 85],
    'Art':     [None, 73, 88, 84, 91, 77, None, 80, 85, None]
}

# Creating a dataframe
x_df = pd.DataFrame(data)

# Displaying the dataframe
x_df
```

```python
# Replacing null values in the student column with unknown
x_df['Student'] = x_df['Student'].fillna('Unknown')

# Replacing null values in the math column with 0
x_df['Math'] = x_df['Math'].fillna(0)

# Filling null values in the englich column using mean
x_df['English'] = x_df['English'].fillna(x_df['English'].mean())

# Filling null values in the science column using median
x_df['Science'] = x_df['Science'].fillna(x_df['Science'].median())

# Filling null values in the history column using forward fill
x_df['History'] = x_df['History'].ffill()

# Filling null values in the art column using backward fill
x_df['Art'] = x_df['Art'].bfill()

# Displaying the resulting dataframe
x_df
```

# Task 5: Concatenation and Merging

* Concatenate df1 and df2 vertically.
* Merge the result with df3 horizontally.
* Ensure the resulting DataFrame has the correct shape and all data.

```python
# Provided dataframe data
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})
```

```python
# merging df_combined with df3
final_df = pd.concat([df_combined, df3], axis=1)
final_df
```

Tuples are immutable meaning that once they are created, they cannot be changed whereas sets ar mutable meaning that you can add or delete elements from them after their creation.

# Task 12: Pandas Data Analysis

* Calculate and display the average salary by department.
* Extract and display the names of employees with salary > 60000.
* Add a 'Bonus' column that is 10% of the salary.
* Ensure all operations are performed on the DataFrame correctly.

```python
# Employee DataFrame
df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})
df_employees
```

```python
# Average salary by department
avg_salary_by_dept = df_employees.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
print(avg_salary_by_dept)
```

```python
# Employees with high salaries
high_earners = df_employees[df_employees['Salary'] > 60000]['Name']
print("\nEmployees with Salary > 60000:")
print(high_earners)
```

```python
# Creating bonus column
df_employees['Bonus'] = df_employees['Salary'] * 0.10
print("\nDataFrame with Bonus Column:")
print(df_employees)
```
